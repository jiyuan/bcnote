# 神经网络原理

### 1. 统计学习理论

通过描述一种学习理论来讨论神经网络的统计特性。以*监督学习*为例。

监督学习模型包括三个部分：
+ 环境。即随机向量$\mathbb{x}$，服从一个未知的概率分布函数。
+ 教师。响应$\mathbb{d}$，$d=f(x,v)$，$v$是噪声。
+ 学习机器(算法)。$y=F(x,w)$，$w$是参数，即权值。

监督学习要解决的问题是：以最优化的方式选择逼近$d$的$F(x,w)$。

监督学习的可行性取决于，训练样本$\{(x_id_i)\}_{i=1}^N$是否包含足够的信息来构建具有良好推广性能的学习机器。

这个问题的答案需要用到Vapnik和Chervonenkis在1971年创造的工具。将监督学习问题视为一个“逼近问题”来讨论，即要寻找$f(x)$的最好的逼近函数$F(x,w)$。

用$L(d,F(x,w))$度量输入向量$x$的期望响应$d$和机器产生的响应$F(x,w)$之间的“差异”。“差异”的期望值由下面的*风险泛函*来定义：

\begin{equation*}
R(w)=\int L(d,F(x,w))d\mathcal{F}_{X,D}(x,d)
\end{equation*}
目标是使$R(w)$最小化。

$R(w)$的求值是复杂的，因为联合分布函数$\mathcal{F}_{X,D}(x,d)$通常是未知的。为克服这一困难，我们采用经验风险最小化归纳原则(Vapnik,1982)。这一原则完全依赖于训练数据集$\mathcal{J}$的可用性，这使得它非常适合于神经网络的设计原理。

___

#### \# 经验风险泛函

给定训练样本集$\mathcal{J}=\{(x_i,d_i)\}_{i=1}^N$，经验风险泛函定义为

\begin{equation*}
R_{emp}(w)=\frac{1}{N} \sum_{i=1}^N L(d_i,F(x_i,w))
\end{equation*}

#### \# 经验风险最小化原则

这一原则的基本思想就是，以经验风险泛函$R_{emp}(w)$代替风险泛函$R(w)$。

要考虑的是，以$R(w)$和$R_{emp}(w)$作为度量时，$F(x,w_{emp})$和$F(x,w_o)$是否“接近”。这一假设成立的充分必要条件需要以下的概率一致收敛，

\begin{equation*}
P(\sup_{w\in W} |R(w)-R_{emp}(w)|>\epsilon) \rightarrow 0  \quad \mbox{当$N\rightarrow\infty$}
\end{equation*}
